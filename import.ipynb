```python
import json

def process_large_json_file(file_path):
    counter = 0
    with open(file_path, 'r', encoding='utf-8') as file:
        for line_number, line in enumerate(file, start=1):
            if counter>2:
                break
            counter +=1
            line = line.strip()
            
            try:
                data = json.loads(line)
                print("w",line, "\n")
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON on line {line_number}: {e}")
                continue  # Skip the malformed line
            

# Use the function with your file path
input_file_path = 'books.json'
process_large_json_file(input_file_path)
```

```python
import re

def clean_json_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write('[\n')  # Start of JSON array
        
        first_line = True
        for line in infile:
            # Remove any extra spaces or newlines
            line = line.strip()
            # Skip empty lines or invalid JSON objects
            if not line or line == "{}":
                continue
            # If it's not the first valid JSON object, add a comma before the next one
            if not first_line:
                outfile.write(',\n')
            outfile.write(line)
            first_line = False
        
        outfile.write('\n]')  # End of JSON array

# Use the function with your file paths
input_file_path = 'authors.json'
output_file_path = 'cleanedfileauthor.json'
clean_json_file(input_file_path, output_file_path)

print(f"Cleaned JSON saved to {output_file_path}")
```
